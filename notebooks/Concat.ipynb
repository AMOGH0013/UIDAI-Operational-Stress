code to concatinate the slices of DATASET


```python
import pandas as pd
import glob

# find all enrolment csv files
enrolment_files = glob.glob("api_data_aadhar_enrolment_*.csv")

print("Enrolment files found:")
for f in enrolment_files:
    print(f)

# read and concatenate
enrolment_dfs = []
for file in enrolment_files:
    df = pd.read_csv(file)
    enrolment_dfs.append(df)

enrolment_full = pd.concat(enrolment_dfs, ignore_index=True)

# basic sanity checks
print("Total rows:", enrolment_full.shape[0])
print("Date range:", enrolment_full["date"].min(), "to", enrolment_full["date"].max())
print("Sample districts:")
print(enrolment_full["district"].value_counts().head())
```
